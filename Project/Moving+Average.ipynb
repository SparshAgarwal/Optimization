```julia
raw = readcsv("stock-prices-final.csv");
u = zeros(90)
for i = 1:90
    u[i] = i
end


y = raw[1,2:91];
y1 = raw[2,2:91];
y2 = raw[3,2:91];
y3 = raw[4,2:91];
y4 = raw[5,2:91];
y5 = raw[6,2:91];
y6 = raw[7,2:91];
y7 = raw[8,2:91];

T = length(u);
```

```julia


using PyPlot,JuMP, Gurobi
PyPlot.svg(true)
m = Model(solver=GurobiSolver(OutputFlag=0))
figure(figsize=(12,4))
plot([y y1 y2 y3 y5 y6 y7],".-");
legend(["input u", "output y"], loc="lower right");
title("Input and output data");




w1 = 5
A = zeros(T,w1)
for i = 1:w1
    A[i:end,i] = u[1:end-i+1]
end
println(A)

wopt = inv(A'*A)*(A'*y)
println(wopt)
yest = A*wopt

figure(figsize=(12,4))
plot(y,"g.-",yest,"m.-",)
legend(["true output", "predicted output"], loc="upper right");
title("Moving averages model");

```
